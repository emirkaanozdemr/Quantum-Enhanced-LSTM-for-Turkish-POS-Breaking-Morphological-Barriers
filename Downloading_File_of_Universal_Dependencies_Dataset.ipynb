```python
import pandas as pd
from conllu import parse


with open("/content/ud-treebanks-v2.15/UD_Turkish-BOUN/tr_boun-ud-train.conllu", "r", encoding="utf-8") as f:
    data = f.read()

sentences = parse(data)

rows = []
for sentence in sentences:
    sentence_text = " ".join([token['form'] for token in sentence])
    upos_sequence = " ".join([token['upos'] for token in sentence])
    rows.append({"Sentence": sentence_text, "UPOS Sequence": upos_sequence})


BOUN_train = pd.DataFrame(rows)

BOUN_train
```
